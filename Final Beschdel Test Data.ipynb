```python
import pandas as pd
```

```python
bttest=pd.read_csv('Complete_beschdel.csv')
bttest
```

```python
boxoffice=pd.read_csv('Boxoffice.csv')
boxoffice
```

```python
new_df=pd.merge(boxoffice, bttest, left_on='Movie Title', right_on='title', how='inner').drop(columns=['title', ])
new_df
```

```python
print(f"Domestic collection value unknown: {new_df['Deomestic'].isin(['-']).sum()}")
print(f"Worldwide collection value unknown: {new_df['World Wide'].isin(['-']).sum()}")
print(f"Foreign collection value unknown: {new_df['Foreign'].isin(['-']).sum()}")
```

```python
new_df.duplicated().sum()
```

```python
new_df.drop_duplicates(inplace=True)
new_df = new_df.reset_index(drop=True)
new_df
```

```python
new_df.drop(columns=['Foreign'], inplace=True)
new_df
```

```python
new_df['World Wide'] = new_df['World Wide'].replace('[\$,]', '', regex=True).astype(float)
new_df['Deomestic'] = new_df['Deomestic'].replace('[\$,]', '', regex=True)  # Remove dollar sign and commas
new_df['Deomestic'] = pd.to_numeric(new_df['Deomestic'], errors='coerce')  # Convert to numeric, replace non-numeric values with NaN
```

```python
new_df.to_csv('Beschdel_final.csv', index=False)
```
